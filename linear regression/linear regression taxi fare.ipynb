```python
# Import necessary libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import pandas as pd
```

```python
# Load the dataset
data = pd.read_csv('C:/Users/Sumanth/Downloads/taxi fare.csv')
data
```

```python
# Exclude non-numeric columns
numeric_data = data.select_dtypes(include=['float64', 'int64'])

# Compute the correlation matrix
correlation_matrix = numeric_data.corr()

# Extract the correlation values with the target variable (e.g., 'price')
correlation_with_price = correlation_matrix['total_fare']

# Identify variables with positive correlation
positive_correlation = correlation_with_price[correlation_with_price > 0]

# Display the variables with positive correlation
print("Variables with Positive Correlation:")
print(positive_correlation)
```

```python
# Selecting only the 'distance_traveled' and 'fare' columns from the DataFrame
data=data[['distance_traveled','fare']]
data
```

```python
# Extracting the 'distance_traveled' column as the feature (X)
X=data[['distance_traveled']]
# Extracting the 'fare' column as the target variable (Y)
Y=data[['fare']]
```

```python
X
```

```python
Y
```

```python
# Split the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
# Create a Linear Regression model
model = LinearRegression()
# Fit the model to the training data
model.fit(X_train, Y_train)
```

```python
# Make predictions on the test data using the trained model
Y_pred_test = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error

# After making predictions (y_pred) and having the actual target values (y_test)
# Calculate R-squared
r2 = r2_score(Y_test, Y_pred_test)

# Calculate mean squared error (MSE)
mse = mean_squared_error(Y_test, Y_pred_test)

# Print the results
print("R-squared:", r2)
print("Mean squared error:", mse)
```
