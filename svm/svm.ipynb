```python
# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Loading your dataset
df=pd.read_csv('C:/Users/Sumanth/Downloads/loan_approval_dataset (1).csv')
df
```

```python
#Drop the null values
df.dropna()
```

```python
#viewing the columns
df.columns
```

```python
# Dropping the column which is not necessary
df = df.drop(['loan_id'],axis=1)
```

```python
#Viewing the data
df
```

```python
# Separate features (X) and target variable (y)
X = df.drop(' loan_status', axis=1) 
y = df[' loan_status']
```

```python
# Split the data into training and testing sets
# Use 80% for training and 20% for testing, with a fixed random state for reproducibility
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
# Standardize the features using StandardScaler
# Create a StandardScaler object
from sklearn.preprocessing import StandardScaler    
st_x= StandardScaler()    
# Fit and transform the training data
X_train= st_x.fit_transform(X_train)   
# Transforming the test data using the same scaler 
X_test= st_x.transform(X_test)       
```

```python
# Creating a Support Vector Machine model with a linear kernel
from sklearn import svm

svm_model = svm.SVC(kernel='linear')
```

```python
# Train the Support Vector Machine model on the training data
svm_model.fit(X_train, y_train)

```

```python
# Make predictions on the test data using the trained SVM model
y_pred = svm_model.predict(X_test)

```

```python
# Print or inspect the predicted labels
y_pred
```

```python
from sklearn.metrics import confusion_matrix, classification_report
# Calculate the confusion matrix
cm = confusion_matrix(y_test, y_pred)
# Print the confusion matrix
print("Confusion Matrix:\n", cm)
```

```python
# Calculate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
accuracy
```
